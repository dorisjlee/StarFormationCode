```python
data = np.loadtxt("p-pointmass.out",skiprows=3)#,usecols = (1,2,3,4))
```

```python
def plot_all(l,x,d,u,p):
    plt.figure()
    plt.xlabel("Position(x)")
    plt.ylabel("Pressure(p)")
    plt.plot(x,p,'o')
    plt.plot(x,p,color="red")
    plt.figure()
    plt.xlabel("Position ")
    plt.ylabel("Velocity")
    plt.plot(x,u,'o')
    plt.plot(x,u,color="red")
    plt.figure()
    plt.xlabel("Position(x)")
    plt.ylabel("Density(d)")
    plt.plot(x,d,'o')
    plt.plot(x,d,color="red")
    plt.figure()
    plt.xlabel("Position")
    plt.ylabel("Level")
    plt.plot(x,l,color="red")
```

```python
l = data[::,0]
x= data[::,1]
d = data[::,2]
u = data[::,3]
p = data[::,4]
plot_all(l,x,d,u,p)
```
