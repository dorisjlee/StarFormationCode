```python
fattening_factor= [1000,100,31.6, 10]
# Time until collapse till point-ish
time = [4.149995E+11, 1.8727E+12,4.5127E+12,1.7015E+13]
# Order of magnitude computational time in seconds
comp_time = [120,3600,5400,28800]
```

```python
plt.loglog(fattening_factor,time)
plt.loglog(fattening_factor,time,'o',color = "red")
plt.xlabel("Fattening Factor ",fontsize=14)
plt.ylabel("Time",fontsize=14)
```

```python
plt.semilogx(fattening_factor,time)
plt.xlabel("Fattening Factor ",fontsize=14)
plt.ylabel("Time",fontsize=14)
```

```python
plt.loglog(fattening_factor,comp_time)
plt.xlabel("Fattening Factor ",fontsize=14)
plt.ylabel("Computational Time",fontsize=14)
```
