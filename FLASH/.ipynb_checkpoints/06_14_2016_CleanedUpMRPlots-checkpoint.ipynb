```python
xi_range = np.logspace(-1.5,1.5,num=30)
tlst = [0,28,30,32,40]
hsv = plt.get_cmap('Blues')
colors = hsv(np.linspace(0.7, 1.0, len(tlst)))
for s , c in zip(tlst,colors):
    data = np.loadtxt("../lev8nosink/compare_new_fast_sum_args_list{}_lev7.txt".format(s))
    plt.loglog(xi_range[::-1],4*np.pi*G*data/c_s,".",label= "t={}".format(s), color=c)
    
min_cell_size =  1e19/2**(9)*1.057E-17
plt.axvline(min_cell_size,c="red")
plt.axvline(4e18*1.057E-17,c="red")

plt.xlabel(r"log $\xi$",fontsize=16)
plt.ylabel(r"log m/$\xi$",fontsize=16)
plt.title("Hi-res no sink",fontsize=16)


xi_range = np.logspace(-1.5,1.5,num=30)
tlst = [0,28,30,32,40]
hsv = plt.get_cmap('Reds')
colors = hsv(np.linspace(0.5, 1.0, len(tlst)))
for s , c in zip(tlst,colors):
    data = np.loadtxt("../low_res_no_sink/compare_new_fast_sum_args_list{}_lev5.txt".format(s))
    plt.loglog(xi_range[::-1],4*np.pi*G*data/c_s,label= "t={}".format(s), color=c)
min_cell_size =  1e19/2**(7)*1.057E-17
plt.axvline(min_cell_size,c="blue")
plt.axvline(4e18*1.057E-17,c="blue")
plt.legend(loc ="center left", bbox_to_anchor=(1, 0.5))
plt.xlabel(r"$\xi$",fontsize=16)
plt.ylabel(r"m/$\xi$",fontsize=16)
# plt.title("New code, better ranges (No Sink)",fontsize=16)
```

The OOM error came from the fact that we were forcing the dr to be the cell size of the low-res grid, while our FRB was still projected onto lev 7. If we project both datasets onto the low-res grid (lev 5) and set dr as lev 5 cell size, then our results converges.

```python
xi_range = np.logspace(-1.5,1.5,num=30)
tlst = [0,28,30,32,40]
hsv = plt.get_cmap('Blues')
colors = hsv(np.linspace(0.7, 1.0, len(tlst)))
for s , c in zip(tlst,colors):
    data = np.loadtxt("../lev8nosink/compare_new_fast_sum_args_list{}_lev5.txt".format(s))
    plt.loglog(xi_range[::-1],4*np.pi*G*data/c_s,".",label= "t={}".format(s), color=c)
    
min_cell_size =  1e19/2**(9)*1.057E-17
plt.axvline(min_cell_size,c="red")
plt.axvline(4e18*1.057E-17,c="red")

plt.xlabel(r"log $\xi$",fontsize=16)
plt.ylabel(r"log m/$\xi$",fontsize=16)
plt.title("Hi-res no sink",fontsize=16)


xi_range = np.logspace(-1.5,1.5,num=30)
tlst = [0,28,30,32,40]
hsv = plt.get_cmap('Reds')
colors = hsv(np.linspace(0.5, 1.0, len(tlst)))
for s , c in zip(tlst,colors):
    data = np.loadtxt("../low_res_no_sink/compare_new_fast_sum_args_list{}_lev5.txt".format(s))
    plt.loglog(xi_range[::-1],4*np.pi*G*data/c_s,label= "t={}".format(s), color=c)
min_cell_size =  1e19/2**(7)*1.057E-17
plt.axvline(min_cell_size,c="blue")
plt.axvline(4e18*1.057E-17,c="blue")
plt.legend(loc ="center left", bbox_to_anchor=(1, 0.5))
plt.xlabel(r"$\xi$",fontsize=16)
plt.ylabel(r"m/$\xi$",fontsize=16)
# plt.title("New code, better ranges (No Sink)",fontsize=16)
```
