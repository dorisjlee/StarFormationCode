#### Derived Fields

Even though density is already computed we want to try deriving it as a field

```python
from yt import derived_field # Remember this ! 
```

So if I want to know how big this simulation is modelling over I could make a rough estimate based on the level0 grid :

```python
ds.print_stats()
print "Each side of the box is minmum {} pcs".format(1.562e-3*128)
```

This makes sense since this is on the order of mag of 0.1 pc for MC

```python
@derived_field(name = "derived_density",units="g/m^3")
def _density(field,data):
    return data["cell_mass"]/data["cell_volume"]
```
