On edison, running for 3 hours 

```python
nproc = np.array([1, 24, 48,120, 240, 360, 480,720, 960, 1920, 2400])
timestep = np.array([22, 88, 66, 87, 114, 140,140,138,138,103, 92])
```

```python
plt.plot(nproc,timestep,'o')
plt.plot(nproc,timestep)
plt.xlabel("nproc")
plt.ylabel("Timestep")
plt.title("Strong Scaling on Edison (3hrs)")
```
